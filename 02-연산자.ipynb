## 1. bmi 계산기
- bmi = 몸무게(kg) / 키(m) 제곱 2
- weight 에 몸무게, height 에 키를 저장하고 bmi 를 계산하기

```python
weight = 50
height = 1.6
bmi = weight / height**2

bmi
```

## 2. 원 넓이와 둘레 계산 (pi 는 3.14 로 계산)
- 원 넓이 = pi x r^2
- 원 둘레 = 2 x pi x r
- r = 3

```python
pi = 3.14
r = 3

넓이 = pi * r**2
둘레 = 2 * pi * r

넓이, 둘레
```

## 3. 초를 시간, 분, 초 로 바꾸기
- 계산을 통해 h, m, s 에 시간, 분, 초로 전환
- seconds = 3675  # 1시간 1분 15초

```python
8//3s
```

```python
s = 1
m = 60*s
h = 60*m
seconds = 3675
hours = seconds//h
minutes = (seconds//m) %60
seconds_value = seconds%60
time = f'{hours}시간 {minutes}분 {seconds_value}초'

hours, minutes, seconds_value, time
```

## 4. 세자리 수 분리하기
- 숫자를 백, 십, 일의 자리로 분리하여 변수에 저장하기
- n = 472

```python
n = 472
h = 100
t = 10
o = 1

hundreds = n//h
ten = n//t - hundreds*10
one = n//o - (hundreds*h) - (ten*t)

hundreds, ten, one
```
