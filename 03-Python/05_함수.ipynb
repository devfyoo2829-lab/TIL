```python
# name: plus_one / in: 숫자 / out: 숫자 +1

def plus_one(num):
    if num%2 == 0:
        return "짝수"
    else:
        return "홀수"

plus_one(10)
```

```python
# n번 인사하는 함수
# name: hello_ntimes

def hello_ntimes( n ):
    # for-매개변수에 들어간 값을 사용하지 않을 때 → _
    for _ in range(n):
        print('안녕')
    

hello_ntimes(5) # 5번 hello를 출력
```

```python
# 3제곱 해주는 함수
# name: cube, in: 숫자 1개, out: 숫자의 세제곱
def cube(n):
    return n**3


cube(3)  #27
```

```python
# 원기둥의 부피를 구하는 함수
# name: cylinder, in: 반지름/높이, out: 부피(pi * r^2 * h)
import numpy as np

def cylinder(r, h):
    return round(np.pi*r**2*h,2)

cylinder(2,5) #반지름 2, 높이 5
```

```python
# 리스트 내부의 숫자들의 평균을 구해주는 함수
# name: get_avg, in: 숫자들로 만들어진 리스트, out: 평균

def get_avg(x):
    return sum(x)/len(x)

get_avg([1,2,3,4,5]) #3
```

```python

# 비밀번호 체크해주는 함수 (숫자 포함, 8글자 이상)
# name: check_pw, in: 비밀번호 문자열, out: True/False


def check_pw(password):
    is_okay = 'fail'
    if len(password) >= 8:
        for char in password:
            if char in '0123456789':
                is_okay = 'pass'
    return is_okay

check_pw('abcd1234') #True
#check_pw('abcdefg') #False
```

```python
# BMI 기준으로, 비만여부 확인해주는 함수 (체중/키(m)^2)
# 남성 27 이상일 경우 비만
# 여성 25 이상일 경우 비만
weight = float(0)
height = float(0)

def check_bmi(weight, height, gender):
    bmi = weight/(height)**2
    if bmi >= 27 and gender == 'M':
        return 'Yes'
    elif bmi >= 25 and gender == 'F':
        return 'Yes'
    return 'No', round(bmi,1)


print(check_bmi(80.0, 1.75, 'M'), check_bmi(60.0, 1.6, 'F'))

```
