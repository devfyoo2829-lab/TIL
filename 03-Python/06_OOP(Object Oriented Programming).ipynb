### OOP(객체지향언어) → 실존하는 값이 주인공
- Object : 객체 = 값 (`1`, `'abc'`, `[1,2,3]`)
- Oriented : 지향 = 핵심, 주인공
- Class - Instance


```python
# Person
# 키, 몸무게, 나이, 이름
# 인사, 공부

# python에서 클래스 이름은 PascalCase로 짓는다
class Person:
    # 1. 특성 (attribute)
    height = 165
    weight = 55
    age = 20
    name = '박제리'
    iq = 130

    # 2. 행동 (method)
    def greeting(self):
        return f'안녕하세요, {self.name}입니다. {self.age}살 입니다.'

    def study(self):
        self.iq += 5 
```

```python
# Person 개념적 대상을 → 실제 생성
p1 = Person()
p2 = Person()
```

```python
# 특성
p1.age = 30
p1.name = '김로봇'
p1.weight = 82
p1.height = 182

# 행동
p1.greeting()
```

```python
# 특성
p2.age = 23
p2.name = '박사람'
p2.weight = 45
p2.height = 155

# 행동
p2.greeting()
```

```python
# 원(Circle)
class Circle:
    # 생성자함수 → 처음 인스턴스가 만들어질 때 자동 호출되는 함수
    def __init__(self, r):
        self.r = r
        print(f'반지름이 {self.r}인 원이 생성되었습니다.')

    def get_area(self):
        print(f'반지름이 {self.r}인 원의 넓이를 구했습니다.')
        return 3.14 * self.r **2

    def get_perimeter(self):
        print(f'반지름이 {self.r}인 원의 둘레를 구했습니다.')
        return 3.14 * 2 * self.r

    def __add__(self, other):
        return Circle(self.r + other.r)
    
c1 = Circle(3) #반지름이 3인 원
c2 = Circle(5) #반지름이 5인 원

print(c1.get_area()) #반지름이 3인 원의 넓이
print(c2.get_area()) #r=5의 넓이

print(c1.get_perimeter())
c3 = c1 + c2 # r=3 + r=5 → r=8 원을 새롭게 만들겠다
c3.get_area()
```
