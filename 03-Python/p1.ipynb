```python
# 문제 1. 점수 평균과 합격 판별

scores = [80, 75, 90, 60, 100]
idx = 0
sum = 0

while idx < len(scores):
    sum += scores[idx]
    idx += 1
print(f'총합 = {sum}')

total = 0
for num in scores:
    total += num
print(f'총합 = {total}')

average = sum / len(scores)
print(f'평균 = {average}')

if average >=80:
    print('당신은 합격입니다.')
else:
    print('당신은 불합격입니다.')
```

```python
# 문제 4. 이름 존재 여부 확인
target = "민수"
names = ["철수", "영희", "민수", "지영"]

if target in names:
    print('존재합니다.')
else :
    print('없습니다.')



result = '없습니다.'

for name in names:
    if name == target:
        result = '있습니다'

print(result)
```

```python
# 문제 6. 재고 확인 시스템
inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}

if inventory['apple'] >= 1:
    print(f"apple : 판매 가능")
elif inventory['apple'] == 0:
    print(f"apple : 품절")

if inventory['banana'] >= 1:
    print(f"banana : 판매 가능")
elif inventory['banana'] == 0:
    print(f"banana : 품절")

if inventory['orange'] >= 1:
    print(f"orange : 판매 가능")
elif inventory['orange'] == 0:
    print(f"orange : 품절")

# 줄여서 하자면
for key in inventory:
    if inventory[key] >= 1:
        print(f"{key} : 판매 가능")
    elif inventory[key] == 0:
        print(f"{key} : 품절")
```

```python
# 문제 7. 리스트 안 딕셔너리 평균 구하기

students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]

sum = students[0]['score']
for i in range(2):
    sum += students[i+1]['score']

print(f'총합 :{sum}')
average = sum/len(students)
print(f'평균 : {average}')

for j in range(3):
    if students[j]['score']>=average:
        print(f'통과한 사람 : {students[j]['name']}')
```
